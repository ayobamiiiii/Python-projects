```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss
from sklearn.metrics import confusion_matrix, accuracy_score
import sklearn.metrics as metrics
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv")
df.head()
```

```python
df.shape
```

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
```

```python
df_sydney_processed.head()
```

```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
```

```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
df_sydney_processed.head()
```

```python
X = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=10)

# Display the split data
print("X_train:\n", X_train)
print("X_test:\n", X_test)
print("y_train:\n", y_train)
print("y_test:\n", y_test)
```

```python
LinearReg = LinearRegression()

# Step 6: Train the model using the training data
LinearReg.fit(X_train, y_train)
```

```python
predictions = LinearReg.predict(X_test)
print("Predicted values:", predictions)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculating the metrics
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

# Printing the metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("R-squared (RÂ²):", r2)
```

```python
Report = pd.DataFrame({
    'Metric': ['Mean Absolute Error (MAE)', 'Mean Squared Error (MSE)', 'R-squared (RÂ²)'],
    'Value': [mae, mse, r2]
})

# Displaying the DataFrame
print(Report)
```

```python
KNN = KNeighborsRegressor(n_neighbors=4)

# Train the KNN model using the training data
KNN.fit(X_train, y_train)

# Predicting the target values for the test set using the trained KNN model
predictions_knn = KNN.predict(X_test)

# Calculating the metrics for the KNN model
mae_knn = mean_absolute_error(y_test, predictions_knn)
mse_knn = mean_squared_error(y_test, predictions_knn)
r2_knn = r2_score(y_test, predictions_knn)

# Creating a DataFrame to display the metrics for both models
metrics_df = pd.DataFrame({
    'Metric': ['Mean Absolute Error (MAE)', 'Mean Squared Error (MSE)', 'R-squared (RÂ²)'],
    'Linear Regression': [mae, mse, r2],
    'KNN (n_neighbors=4)': [mae_knn, mse_knn, r2_knn]
})

# Displaying the DataFrame
print(metrics_df)
```

```python
from sklearn.metrics import accuracy_score, jaccard_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Encoding the categorical target variable
le = LabelEncoder()
y_encoded = le.fit_transform(Y)

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=10)

# Create the KNN model with n_neighbors set to 4
KNN = KNeighborsClassifier(n_neighbors=4)

# Train the KNN model using the training data
KNN.fit(X_train, y_train)

# Predicting the classes for the test set using the trained KNN model
predictions_knn = KNN.predict(X_test)

# Calculating the metrics for the KNN model
KNN_Accuracy_Score = accuracy_score(y_test, predictions_knn)
KNN_JaccardIndex = jaccard_score(y_test, predictions_knn, average='macro')
KNN_F1_Score = f1_score(y_test, predictions_knn, average='macro')

# Creating a DataFrame to display the metrics
metrics_df = pd.DataFrame({
    'Metric': ['Accuracy Score', 'Jaccard Index', 'F1 Score'],
    'KNN (n_neighbors=4)': [KNN_Accuracy_Score, KNN_JaccardIndex, KNN_F1_Score]
})

# Displaying the DataFrame
print(metrics_df)
```

```python
Tree = DecisionTreeClassifier(random_state=10)

# Train the Decision Tree model using the training data
Tree.fit(X_train, y_train)

# Predicting the classes for the test set using the trained Decision Tree model
predictions_tree = Tree.predict(X_test)

# Calculating the metrics for the Decision Tree model
Tree_Accuracy_Score = accuracy_score(y_test, predictions_tree)
Tree_JaccardIndex = jaccard_score(y_test, predictions_tree, average='macro')
Tree_F1_Score = f1_score(y_test, predictions_tree, average='macro')

# Creating a DataFrame to display the metrics
metrics_df = pd.DataFrame({
    'Metric': ['Accuracy Score', 'Jaccard Index', 'F1 Score'],
    'Decision Tree': [Tree_Accuracy_Score, Tree_JaccardIndex, Tree_F1_Score]
})

# Displaying the DataFrame
print(metrics_df)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
LR = LogisticRegression(solver='liblinear')

# Train the Logistic Regression model using the training data
LR.fit(X_train, y_train)

# Use the model to make predictions on the testing data
predictions = LR.predict(X_test)

# Use the model to predict probabilities on the testing data
predict_proba = LR.predict_proba(X_test)

# Display the predictions and the predicted probabilities
print("Predictions:", predictions)
```

```python
print("Predicted Probabilities:", predict_proba)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, log_loss

LR_Accuracy_Score = accuracy_score(y_test, predictions)

precision = precision_score(y_test, predictions, average='weighted')

recall = recall_score(y_test, predictions, average='weighted')

LR_F1_Score = f1_score(y_test, predictions, average='weighted')

LR_JaccardIndex = jaccard_score(y_test, predictions, average='weighted')

LR_log_loss = log_loss(y_test, predict_proba)

metrics_df = pd.DataFrame({
    'Metric': ['LR_Accuracy_Score', 'Precision', 'Recall', 'LR_F1_Score', 'LR_JaccardIndex', 'LR_log_loss'],
    'Value': [LR_Accuracy_Score, precision, recall, LR_F1_Score, LR_JaccardIndex, LR_log_loss]
})


print(metrics_df)
```
